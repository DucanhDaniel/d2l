### Mạng nơ ron hồi tiếp sâu
- Ta mới chỉ thảo luận về các mạng nơ ron hồi tiếp với duy nhất một tầng ẩn đơn hướng. Trong đó, cách các biến tiềm ẩn và các quan sát tương tác với nhau còn khá tùy ý. 
- Đây là một vấn đề thách thức với các mạng đơn tầng, trong trường hợp của perceptron, ta giải quyết vấn đề này bằng cách đưa thêm nhiều tầng vào trong mạng. Cách này hơi phức tạp đối với trường hợp của mạng RNN, vì __đầu tiên chúng ta cần phải quyết định thêm tính phi tuyến vào mạng ở đâu và như thế nào.__
    + Ta có thể bổ sung tính phi tuyến vào các cơ chế cổng. __Tức là, thay vì sử dụng một tầng perceptron duy nhất, ta có thể sử dụng nhiều tầng perceptron.__ Cách này không làm thay đổi cơ chế của mạng LSTM, ngược lại làm cho nó tinh xảo hơn. Điều này chỉ có lợi nếu ta tin rằng LSTM biểu diễn một hình thái phổ quát nào đó về cách hoạt động của các mô hình tự hồi quy biến tiềm ẩn.
    + Ta có thể __chồng nhiều tầng LSTM lên nhau.__ Cách này tạo ra một cơ chế linh hoạt hơn nhờ vào sự kết hợp giữa các tầng đơn giản. Đặc biệt là, các đặc tính liên quan của dữ liệu có thể được biểu diễn ở các tầng khác nhau. Ví dụ, ta có thể muốn lưu dữ liệu về tình hình thị trường tài chính ở tầng cao hơn trong khi đó chỉ ghi lại động lực thời hạn ngắn hơn ở một tầng thấp hơn.

- Hình dưới mô tả một mạng nơ ron hồi tiếp sâu với L tầng ẩn. Mỗi trạng thái ẩn liên tục được truyền tới bước thời gian kế tiếp ở tầng hiện hiện tại và tới bước thời gian hiện tại ở tầng kế tiếp. 

    

#### 9.3.1 Các Phụ thuộc Hàm
- Tại bước thời gian t, giả sử rằng chúng ta có một minibatch X_t (n x d) với n là số lượng mẫu, d là số lượng đầu vào. Trạng thái của tầng ẩn l là  với h là số đơn vị ẩn. Biến tầng đầu ra là O_t (n x q) với q là số lượng đầu ra và một hàm kích hoạt tầng ẩn f_l cho tầng l. 
- Ta tính toán trạng thái ẩn của tầng đầu tiên như trước đây, sử dụng đầu vào là X_t. Đối với tất cả các tầng tiếp theo, trạng thái ẩn của tầng trước được sử dụng thay cho X_t

    

- Cuối cùng, tầng đầu ra chỉ dựa trên trạng thái ẩn của tầng ẩn L. Chúng ta sử dụng một hàm đầu ra g để xử lý trạng thái này.

    

- Giống như MLP, số tầng ẩn L và số đơn vị ẩn h được gọi là các siêu tham số. Đặc biệt, chúng ta có thể chọn một trong các kiến trúc RNN, GRU, LSTM để xây dựng mô hình.
