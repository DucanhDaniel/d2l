### The Viterbi Algorithm



### Initialization step
- Trong bước đầu tiên, tính ma trận C và D.
Ma trận C cho biết xác suất để từ state i sang word w_j.

Ma trận D cho biết state trước đó khi gán state i cho word w_j.


### Viterbi: Forward pas

Công thức ma trận C:
C(i, j) -> i là chỉ số tag, j là chỉ số word w_j
C(i, j) bằng xác suất lớn nhất của công thức sau:
- k là chỉ số tag
- C(k, j - 1) là xác suất để đi từ tag k đến từ j - 1
- a(k, i) là xác suất để đi từ tag k đến tag i hiện tại
- b(i, cindex(w_j)) là xác suất để đi từ tag i hiện tại đến word w_j
* d(i, j) cho biết nếu chọn tag i cho w_j thì trước đó tag k sẽ được gán cho w_(j - 1)
