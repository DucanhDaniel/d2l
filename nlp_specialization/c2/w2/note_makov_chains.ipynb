### Hidden Markov Models
- Với mô hình markov, ma trận transition cho biết xác suất từ 1 POS sang 1 POS khác. 
- Mô hình markov ẩn sử dụng emmission probabilities cho biết xác suất từ 1 tag ra một từ.




- Note: Không được smoothing cho initial state, do làm như vậy sẽ khiến cho tất cả các POS tag đều có thể được gắn sau initial tag.
